## Normalisation

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('Sales_Transactions_Dataset_Weekly.csv',index_col=0)
data.head()
```

```python
def z_score_norm(a):
	a = ( a - np.mean(a)) / np.std(a)
	return a
z_score_norm(data).head()
```

```python
def min_max_norm(a):
	mn = np.min(a)
	mx = np.max(a)
	new_mn = 1
	new_mx = 5
	a = (a - mn) / (mx - mn)
	a = a  * (new_mx - new_mn) + new_mn
	return a
min_max_norm(data).head()
```

## Apriori - Cars Data Set

```python
data = pd.read_csv('car_data.csv')

data = data.apply(lambda x: x.name + '_' + x , axis=0)

# data.apply(lambda x: print(x.shape), axis=1)
```

```python
from collections import Counter

data = np.array(data)

li = []
for row in data:
	# row = row[0]
	li.append(row)

c = Counter()
for row in li:
	for item in row:
		s = {item}
		c[frozenset(s)] += 1
print(c)
```

```python
c = list(dict(c).items())
c.sort(key =  lambda x : -x[1])
for i in range(len(c)):
    c[i] = (set(c[i][0]) , c[i][1])
```

```python
def merge(c,data,min_support_count):
	items = c

	n = len(items)
	new_items = Counter()
	sets = []
	for i in range(n):
		for j in range(i+1 , n):
			# count = 0
			s = set(items[i][0])
			s = s.union(set(items[j][0]))
			sets.append(frozenset(s))
	sets = list(set(sets))
	sets = [set(s) for s in sets]
	# print(sets)
	for it in data:
		for s in sets:
			it = set(it)
			# print(s , it , s.issubset(it))
			if s.issubset(it):
				new_items[frozenset(s)] += 1
	
	new_items = sorted(list(dict(new_items).items()), key = lambda x : -x[1])
	ret_items = []
	for item in new_items:
		if item[1] >= 3:
			ret_items.append((set(item[0]),item[1]))
	return ret_items
```
